```python
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
formula = 'Response ~ C(Dentist) + C(Method) + C(Dentist):C(Method)'
model = ols(formula, Alloy1_Data).fit()
anova_table = anova_lm(model)
print(anova_table)
```
